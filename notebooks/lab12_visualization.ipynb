## Visualization Choices Documentation

The visualization choices were based on the type of data and the question being answered.

Bar charts were used for categories, cities, countries, and website availability because these variables are discrete and the goal is comparison. Horizontal bar charts were used when labels were longer, making them easier to read.

Histograms were used for longitude and latitude because these are numeric continuous variables. Histograms are appropriate for showing distributions and concentration of values.

Scatter plots were used for longitude and latitude together because they show the relationship between two continuous variables. This makes them suitable for showing geographic spread.

The pie chart was used only for website availability because it is a simple part-to-whole comparison with two categories. For larger category sets, bar charts were preferred because they are easier to compare accurately.

Dashboard layouts were used to combine the most important insights into one view. This supports fast interpretation and summary reporting.

Interactive Plotly charts were used because they support hover tooltips, zooming, legend interaction, and standalone HTML export. This makes the visualizations easier to explore without needing a running Python server.

These choices follow Tufte's principles of graphical excellence by maximizing useful data ink, avoiding chartjunk, using simple encodings, and matching each chart type to the structure of the data.
